```python
import json
import time
import pandas as pd
from kafka import KafkaProducer

# Load configuration from the JSON file
with open('config.json', 'r') as config_file:
    config = json.load(config_file)

# Kafka connection details
kafka_config = config['kafka']
bootstrap_servers = kafka_config['bootstrap_servers']
topic = kafka_config['topic']

# CSV file path
csv_path = "New_Information.csv"  # You can also load CSV path from the config if needed

# Function to serialize data into JSON format
def json_serializer(data):
    return json.dumps(data).encode("utf-8")

if __name__ == "__main__":

    # Load the data from CSV
    data = pd.read_csv(csv_path)
    json_data = data.to_dict(orient='records')
    
    # Create Kafka producer
    producer = KafkaProducer(bootstrap_servers=bootstrap_servers, value_serializer=json_serializer)
    
    # Continuously send data to Kafka topic
    while True:
        for record in json_data:
            print(record)
            producer.send(topic, record)
            time.sleep(5)
```
